```python
import numpy as np

def generateData(data, X_len, Y_len, sensor_num):
    point_num = data.shape[0]  # 总时间点数
    sample_num = point_num - X_len - Y_len + 1  # 生成的样本数
    
    # 使用 numpy 创建空的输入和输出数组
    X = np.zeros((sample_num, X_len, sensor_num))  # 保存输入数据
    Y = np.zeros((sample_num, Y_len, sensor_num))  # 保存输出数据
    
    # 遍历生成每个样本的输入和输出
    for i in range(sample_num):
        # 提取前 X_len 个时间点的数据作为输入 X
        X[i] = data.iloc[i:i+X_len, :sensor_num].values  # 获取前 X_len 行的所有传感器数据
        
        # 提取接下来的 Y_len 个时间点的数据作为输出 Y
        Y[i] = data.iloc[i+X_len:i+X_len+Y_len, :sensor_num].values  # 获取 Y_len 行的所有传感器数据
    
    return X, Y

# 生成任务1和任务2的数据集
X_t1, Y_t1 = generateData(data, 30, 1, sensor_num)  # 任务1数据集 (30个时间步长输入，1个时间步长输出)
X_t2, Y_t2 = generateData(data, 30, horizon, sensor_num)  # 任务2数据集 (30个时间步长输入，5个时间步长输出)

# 输出数据形状
print('任务1数据集输入数据形状：{0}，输出数据形状：{1}'.format(X_t1.shape, Y_t1.shape))
print('任务2数据集输入数据形状：{0}，输出数据形状：{1}'.format(X_t2.shape, Y_t2.shape))
```

```python
class MultiTimeSeriesDataset(): #定义MultiTimeSeriesDataset类
    def __init__(self, X, Y): #构造方法
        self.X, self.Y = X, Y #设置输入数据和输出数据
    def __len__(self):
        return len(self.X) #获取数据的长度
    def __getitem__(self, index):
        return self.X[index], self.Y[index] #根据索引值为index的数据
```

```python
from mindspore.dataset import GeneratorDataset
def generateMindsporeDataset(X, Y, batch_size=32):
    """
    生成 MindSpore 数据集对象，并将其按批次进行划分
    :param X: 输入数据 (numpy 数组或 pandas DataFrame)
    :param Y: 输出数据 (numpy 数组或 pandas DataFrame)
    :param batch_size: 每批次的大小
    :return: 返回创建的 MindSpore 数据集
    """
    # 将输入的 X 和 Y 数据转换为 float32 类型
    X = X.astype(np.float32)
    Y = Y.astype(np.float32)
    # 创建 MultiTimeSeriesDataset 对象
    dataset = MultiTimeSeriesDataset(X, Y)
    # 创建 GeneratorDataset 对象，指定列名
    def generator():
        for i in range(len(dataset)):
            x_data, y_data = dataset[i]
            yield x_data, y_data
    mindspore_dataset = GeneratorDataset(generator, column_names=["data", "label"])   
    # 批处理数据
    mindspore_dataset = mindspore_dataset.batch(batch_size=batch_size, drop_remainder=False)   
    return mindspore_dataset
```

```python
import mindspore
mindspore.set_context(mode=mindspore.GRAPH_MODE)
import numpy as np
from mindspore import Tensor

class MODEL_RUN:
    def __init__(self, model, loss_fn, optimizer=None, grad_fn=None, lr_scheduler=None, 
                 early_stopping=True, patience=5, min_delta=0.0):
        """
        初始化训练模型的类。

        :param model: 待训练的模型
        :param loss_fn: 损失函数
        :param optimizer: 优化器
        :param grad_fn: 用于计算梯度的函数
        :param lr_scheduler: 学习率调度器（可选）
        :param early_stopping: 是否启用早停
        :param patience: 早停耐心值，当验证集无改善次数超过该值时停止
        :param min_delta: 最小改善值，当验证集损失减小的幅度小于该值时，视为无改善
        """
        self.model = model
        self.loss_fn = loss_fn
        self.optimizer = optimizer
        self.grad_fn = grad_fn
        self.lr_scheduler = lr_scheduler
        self.early_stopping = early_stopping
        self.patience = patience
        self.min_delta = min_delta

    def _train_one_step(self, data, label):
        (loss, _), grads = self.grad_fn(data, label)  # 计算损失和梯度
        self.optimizer(grads)  # 执行优化器更新模型参数
        return loss

    def _train_one_epoch(self, train_dataset):
        self.model.set_train(True)
        for data, label in train_dataset.create_tuple_iterator():
            self._train_one_step(data, label)
        # 更新学习率（如果有学习率调度器）
        if self.lr_scheduler:
            self.optimizer.learning_rate = self.lr_scheduler.get_lr()

    def evaluate(self, dataset, step=None):
        self.model.set_train(False)
        ls_pred, ls_label = [], []
        
        # 假设你在其他地方已定义了PV_index
        # 如果没有，需要在类外部给定PV_index或者在类初始化时传入
        # 在此以PV_index为已定义变量为例
        for data, label in dataset.create_tuple_iterator():
            pred = self.model(data)
            ls_pred += list(pred[:,:,PV_index].asnumpy())
            ls_label += list(label[:,:,PV_index].asnumpy())
        return self.loss_fn(Tensor(ls_pred), Tensor(ls_label)), np.array(ls_pred), np.array(ls_label)

    def train(self, train_dataset, val_dataset, max_epoch_num, ckpt_file_path):
        min_loss = Tensor(np.float32('inf'))
        no_improve_count = 0  # 用于统计连续无改善的Epoch数
        print('开始训练......')
        for epoch in range(1, max_epoch_num + 1):
            print(f'第{epoch}/{max_epoch_num}轮')
            self._train_one_epoch(train_dataset)

            # 计算训练集和验证集的损失
            train_loss, _, _ = self.evaluate(train_dataset)
            eval_loss, _, _ = self.evaluate(val_dataset)
            print(f'训练集损失：{train_loss}, 验证集损失：{eval_loss}')

            # 检查是否有改善
            if (min_loss - eval_loss) > self.min_delta:
                # 如果验证集loss有显著改善，更新最小loss和模型
                mindspore.save_checkpoint(self.model, ckpt_file_path)
                min_loss = eval_loss
                no_improve_count = 0
            else:
                no_improve_count += 1

            # 判断是否需要早停
            if self.early_stopping and no_improve_count >= self.patience:
                print("验证集损失在连续{}个epoch未改善，触发早停!".format(self.patience))
                break

        print('训练完成！')

    def test(self, test_dataset, ckpt_file_path):
        mindspore.load_checkpoint(ckpt_file_path, net=self.model)
        loss, preds, labels = self.evaluate(test_dataset)
        return loss, preds, labels
```

```python
import mindspore
mindspore.set_context(mode=mindspore.GRAPH_MODE)
import numpy as np
from mindspore import Tensor

class MULTI_STEP_MODEL_RUN(MODEL_RUN):
    def __init__(self, model, loss_fn, optimizer=None, grad_fn=None, lr_scheduler=None, 
                 early_stopping=True, patience=5, min_delta=0.0):
        """
        多步预测模型运行类，继承自已加入早停机制的MODEL_RUN。
        
        :param model: 模型实例
        :param loss_fn: 损失函数
        :param optimizer: 优化器
        :param grad_fn: 梯度计算函数
        :param lr_scheduler: 学习率调度器
        :param early_stopping: 是否启用早停
        :param patience: 早停耐心值
        :param min_delta: 早停时最小改善阈值
        """
        super().__init__(model, loss_fn, optimizer, grad_fn, lr_scheduler, 
                         early_stopping=early_stopping, 
                         patience=patience, 
                         min_delta=min_delta)

    def evaluate(self, dataset):
        """
        对数据集进行多步预测评估。

        注意：本方法使用名为PV_index和OP_index的全局或外部定义的索引，以及horizon作为多步预测的步数。
        在实际代码中需要先定义这几个变量或在类初始化时传入。
        
        :param dataset: 测试/验证数据集
        :return: 返回损失值、预测值、标签值
        """
        self.model.set_train(False)  # 设置为测试模式
        ls_pred, ls_label = [], []  # 用于保存预测结果和标签

        for data, label in dataset.create_tuple_iterator():  # 遍历每批数据
            # muti_step_pred用于保存该批数据的多步预测结果
            muti_step_pred = mindspore.numpy.zeros_like(label[:, :, PV_index])  
            x = data  # 初始输入数据

            for step in range(horizon):  # 对每个时间步进行预测
                pred = self.model(x, step)  # 使用模型进行预测
                # 将当前step的预测结果（特定PV_index维度）存入muti_step_pred对应位置
                muti_step_pred[:, step:step+1, :] = pred[:, :, PV_index]

                concat_op = mindspore.ops.Concat(axis=1)
                # 使用预测结果更新输入：移除最早的时间步，将预测结果拼接到末尾形成下一时刻输入
                x = concat_op((x[:, 1:, :], pred))
                # 对OP控制变量使用真实数据（防止预测误差累积）
                x[:, -1:, OP_index] = label[:, step:step+1, OP_index]

            ls_pred += list(muti_step_pred.asnumpy())  # 保存预测结果
            ls_label += list(label[:, :, PV_index].asnumpy())  # 保存真实标签

        loss_value = self.loss_fn(Tensor(ls_pred), Tensor(ls_label))
        return loss_value, np.array(ls_pred), np.array(ls_label)
```

```python
import mindspore 
from mindspore import nn
class TCN_MLP_with_Bias_Block(nn.Cell):
    def __init__(self): #构造方法
        super().__init__() #调用父类的构造方法
        #比MULTI_STEP_TCN_MLP类增加一个偏差块
        self.bias_block = nn.SequentialCell(
            nn.Dense(sensor_num, 64),
            nn.ReLU(),
            nn.Dense(64, 32),
            nn.ReLU(),
            nn.Dense(32, sensor_num),
            nn.ReLU(),
            nn.Conv2d(in_channels=1, out_channels=1, kernel_size=(32,1), pad_mode='valid')
        )
        #对预测时间步数据的嵌入编码操作
        self.step_embedding = nn.Embedding(horizon, sensor_num)
        #对不同传感器的数据做融合（提取传感器数据间的关联特征）
        self.spatial_mlp = nn.SequentialCell(
            nn.Dense(sensor_num, 128),
            nn.ReLU(),
            nn.Dense(128, 64),
            nn.ReLU(),
            nn.Dense(64, 32),
            nn.ReLU(),
            nn.Dense(32, sensor_num)
        )
        #对时间序列做卷积（提取时间点数据间的关联特征）
        self.tcn = nn.SequentialCell(
            nn.Conv2d(in_channels=1, out_channels=1, kernel_size=(3,1), pad_mode='valid'),
            nn.Conv2d(in_channels=1, out_channels=1, kernel_size=(3,1), pad_mode='valid'),
        )
        #通过一个卷积层得到最后的预测结果
        self.final_conv = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=(26, 1), pad_mode='valid') #使用26*1卷积核，不补边
    def construct(self, x, iter_step): #construct方法
        h = self.spatial_mlp(x) #经过spatial_mlp空间处理后，得到的数据h的形状：[batch_size, 30, 23]
        #输入数据x的形状：[batch_size, 30, 23]
        h = x + h #残差连接，将x和h对应元素相加，得到的数据h的形状：[batch_size, 30, 23]
        h = h.unsqueeze(1) #根据卷积操作需要，将3维数据升为4维数据：[batch_size, 1, 30, 23]
        h = self.tcn(h) #经过tcn时间卷积后，得到的数据x的形状：[batch_size, 1, 26, 23]
        y = self.final_conv(h) #通过26*1的卷积操作后，得到的数据y的形状：[batch_size, 1, 1, 23]
        y = y.squeeze(1) #将前面增加的维度去掉，得到的数据y的形状：[batch_size, 1, 23]
        #计算时间步数据的嵌入编码
        iter_step_tensor = mindspore.numpy.full((x.shape[0], 1), iter_step, dtype=mindspore.int32)
        step_embedding = self.step_embedding(iter_step_tensor) #step_embedding的形状：[batch_size,1,23]
        #将输入数据x与时间步数据的嵌入编码拼接
        concat_op = mindspore.ops.Concat(axis=1)
        bias_input = concat_op((x, step_embedding,y)) #bias_input的形状：[batch_size,32,23]
        bias_output = self.bias_block(bias_input.unsqueeze(1)) #[batch_size, 1, 1, 23]
        bias_output = bias_output.squeeze(1) # [batch_size, 1, 23]
        #加上偏差块的预测结果
        y = y + bias_output #y的形状：[batch_size, 1, 23]
        return y #返回计算结果
```

```python
from mindspore import Tensor
tcn_mlp_bias = TCN_MLP_with_Bias_Block() #创建TCN_MLP_with_Bias_Block类对象tcn_mlp_bias
loss_fn = nn.MAELoss() #定义损失函数

multi_step_optimizer = nn.Adam(tcn_mlp_bias.trainable_params(), 1e-3) #使用Adam优化器

def multi_step_forward_fn(data, label): #定义多步预测前向计算的multi_step_forward_fn方法
    muti_step_pred = mindspore.numpy.zeros_like(label[:,:,PV_index+DV_index])
    x = data
    for step in range(horizon):
        pred = tcn_mlp_bias(x, step) #使用tcn_mlp_bias模型进行预测
        muti_step_pred[:,step:step+1,:] = pred[:,:,PV_index+DV_index] #将当前时间步的预测结果保存到multi_step_pred中
        concat_op = mindspore.ops.Concat(axis=1)
        x = concat_op((x[:,1:,:], pred)) #将预测结果加到输入中
        x[:,-1:,OP_index] = label[:,step:step+1,OP_index] #OP控制变量无法预测、始终使用真实值
    loss = loss_fn(muti_step_pred, label[:,:,PV_index+DV_index]) #根据损失函数计算PV和DV变量的损失值
    return loss, muti_step_pred #返回损失值和预测结果

multi_step_grad_fn = mindspore.value_and_grad(multi_step_forward_fn, None, multi_step_optimizer.parameters, has_aux=True) #获取用于计算梯度的函数
multi_step_model_run = MULTI_STEP_MODEL_RUN(
    model=tcn_mlp_bias,
    loss_fn=loss_fn,
    optimizer=multi_step_optimizer,
    grad_fn=multi_step_grad_fn, 
    early_stopping=True,          # 启用早停机制
    patience=5,                   # 设置早停耐心值为5
    min_delta=0.0001                 # 设置最小改善阈值为0.0001
)
multi_step_model_run.train(train_dataset_t2, val_dataset_t2, 50, 'tcn_mlp_bias_new_with_early_stopping.ckpt') #调用model_run.train方法完成训练
```
